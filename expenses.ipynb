```python
tot=df["Amount"].sum()
print("Monthly Total=", tot)
cat=df.groupby("Category")["Amount"].sum()
print("Category wise total:",cat)
maxi=df["Amount"].max()
print("Single max transaction:", maxi)
avg=df.groupby("Date")["Amount"].sum().mean()
print("Average Daily expense:", avg)
high=df.groupby("Date")["Amount"].sum().idxmax()
print("Highest spending day:",high)
df.groupby("Category")["Amount"].sum().plot(kind="bar")
plt.show()
```
